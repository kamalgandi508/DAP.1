```python
import tensorflow as tf

#Define a constant
a=tf.constant('hello world')
print(a)
print(a.numpy())
```

```python
#Scalar constants
a=tf.constant(42)
b=tf.constant(3.14)
c=tf.constant("Hello,Tensorflow!")
print(a.numpy(),b.numpy(),c.numpy())
```

```python
#vector constants
a=tf.constant([1,2,3])
b=tf.constant([1.0,2.0,3.0])
print(a)
print(a.numpy())
print(b)
print(b.numpy())
```

```python
#martix constants
a=tf.constant([[1,2],[3,4]])
b=tf.constant([[1.0,2.0,],[3.0,4.0]])
print(a)
print(a.numpy())
print(b)
print(b.numpy())
```

```python
#Higher
a=tf.constant([[[1,2,0],[3,4,0]],[[5,6,0],[7,8,0]]])
print(a)
print(a.numpy())
```

```python
#Boolen constants
a=tf.constant(True)
print(a)
print(a.numpy())
```

```python
#special constants
a=tf.zeros(shape=(2,2))
b=tf.ones(shape=(3,3))
print(a)
print(a.numpy())
print(b)
print(b.numpy())
```

```python
import tensorflow as tf
tf.config.run_functions_eagerly(True)
a=tf.constant(5.0)
b=tf.constant(3.0)
c=tf.multiply(a,b)
result=a*b
print(result)
print(result.numpy())
```

```python
a=tf.constant([[1, 2],[3, 4]])
b=tf.constant([[1,2],[3,4]])
c=tf.matmul(a,b)
result = a @ b
print(result)
print(result.numpy())
```

```python
import tensorflow as tf
a = tf.constant(1.25)
print(a)
print(a.numpy())
b=tf.cast(a,dtype=tf.int32)
print(b)
print(b.numpy())
```

```python
import tensorflow as tf
x=tf.constant([[1.,2.,3.],[4.,5.,6.]])
e=tf.transpose(x)
print(e.numpy())

r=tf.reduce_sum(x)
print(r.numpy())
```

```python
f=tf.concat([x,x,x], axis=0)
print(f.numpy())
```

```python
f=tf.concat([x,x,x], axis=1)
print(f.numpy())
```
