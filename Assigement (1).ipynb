```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming the dataset is in a CSV file named 'temperature_data.csv'
df = pd.read_csv('/content/city_temperature 2.csv')

# Calculate average temperature
average_temperature = np.mean(df['AvgTemperature'])

# Find the highest and lowest temperatures
highest_temperature = np.max(df['AvgTemperature'])
lowest_temperature = np.min(df['AvgTemperature'])

# Set your threshold value, for example, 30 degrees Celsius
threshold = 50
# Count the number of days where the temperature exceeded the threshold
days_exceeding_threshold = np.sum(df['AvgTemperature'] > threshold)

# Display the insights
print(f'Average Temperature: {average_temperature:.2f}째C')
print(f'Highest Temperature: {highest_temperature}째C')
print(f'Lowest Temperature: {lowest_temperature}째C')
print(f'Days Exceeding {threshold}째C: {days_exceeding_threshold}')

# Plotting the temperature data
plt.plot(df['Day'], df['AvgTemperature'])
plt.xlabel('Day')
plt.ylabel('Temperature (째C)')
plt.title('Daily Temperatures')
plt.show()
```
